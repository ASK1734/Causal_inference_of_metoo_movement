```python
import pandas as pd
import time 


pd.set_option('display.max_columns', None)
```

```python
S = pd.read_excel('victim_list_victim1.xlsx', sheet_name='S')
T = pd.read_excel('victim_list_victim1.xlsx', sheet_name='T')
A = pd.read_excel('victim_list_victim1.xlsx', sheet_name='Ait')
sumAit = pd.read_excel('victim_list_victim1.xlsx', sheet_name='sumAit')
lagRi = pd.read_excel('victim_list_victim1.xlsx', sheet_name='lagRi')
lagRj = pd.read_excel('victim_list_victim1.xlsx', sheet_name='lagRj for regression')


# victim = df1['victim'].drop_duplicates()
# victim = victim.reset_index().drop(['index'], axis=1)
# victim = victim['victim'].tolist() 
```

```python
target_time_zone = ['2016-08-01 00:00:00', '2016-09-01 00:00:00', 
                    '2016-10-01 00:00:00', '2016-11-01 00:00:00',
                    '2016-12-01 00:00:00', '2017-01-01 00:00:00', 
                    '2017-02-01 00:00:00', '2017-03-01 00:00:00', 
                    '2017-04-01 00:00:00', '2017-05-01 00:00:00',
                    '2017-06-01 00:00:00', '2017-07-01 00:00:00', 
                    '2017-08-01 00:00:00', '2017-09-01 00:00:00', 
                    '2017-10-01 00:00:00', '2017-11-01 00:00:00',
                    '2017-12-01 00:00:00', '2018-01-01 00:00:00', 
                    '2018-02-01 00:00:00', '2018-03-01 00:00:00', 
                    '2018-04-01 00:00:00', '2018-05-01 00:00:00',
                    '2018-06-01 00:00:00', '2018-07-01 00:00:00']
```

```python
df = pd.DataFrame()

for time in target_time_zone:
    # Convert string time to datetime object
    datetime_time = pd.to_datetime(time)

    # Extract data for the given time from each DataFrame and create a temporary DataFrame
    temp_df = pd.DataFrame({
        'id': S['victim_id'],
        't':datetime_time,
        'Sit': S[datetime_time],  # Replace 'value' with the name of the relevant column in S
        'Tit': T[datetime_time],  # Replace 'value' with the name of the relevant column in T
        'Ait': A[datetime_time],  # Replace 'value' with the name of the relevant column in Ait
        'sumAit': sumAit[datetime_time],
        'lagRi': lagRi[datetime_time],
        'lagRj': lagRj[datetime_time],
        'industry': S['section']
    })

    # Concatenate the temporary DataFrame with the main DataFrame
    df = pd.concat([df, temp_df], ignore_index=True)

df['lagRi'] = pd.to_numeric(df['lagRi'], errors='coerce')

#df.to_csv("/Users/chenshaokai/Desktop/工作/中研院/me_too/Alan/google trend & victim list/rawdata/panel_data.csv")
```
